```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

Change the eps and min_samples here, you could get different solutions at each time.

```python
X = pd.read_csv("data/dataset0.csv", header = None)# Let set dataset0.csv as an example
#change the directory of csv file, you could get different results.
dbscan = DBSCAN(eps=0.2, min_samples=10, metric = 'euclidean')
```

```python
plt.figure(figsize=(8, 8))
X.plot.scatter(x=1, y=2)
plt.xlabel('x')
plt.ylabel('y')
mkft1 = plt.show()
```

```python
X1 = X.drop(axis = 1, columns = 0 )
X1 = X1.drop(axis = 1, columns = 3)
X1.to_numpy()
X1
```

```python
dbscan.fit(X1)
```

```python
dbscan.labels_[:5]
```

```python
len(dbscan.core_sample_indices_)
```

```python
x_plot = list(X1[1])
y_plot = list(X1[2])
```

```python
import matplotlib.pyplot as plt
from matplotlib  import cm
import numpy as np
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.set_title("X vs Y AVG",fontsize=14)
ax.set_xlabel("XAVG",fontsize=12)
ax.set_ylabel("YAVG",fontsize=12)
ax.grid(True,linestyle='-',color='0.75')


# scatter with colormap mapping to X_4 value
ax.scatter(x_plot,y_plot,s=20,c=dbscan.labels_, marker = 'o', cmap = cm.jet );

plt.show()
```
